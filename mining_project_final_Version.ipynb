```python
#made by Yusuf İlteriş Kaan Günay & Sultan Melda İnat
# encoding categorical columns
categorical_columns = data.select_dtypes(include=['object']).columns
label_encoders = {}

for column in categorical_columns:
    le = LabelEncoder()
    data[column] = le.fit_transform(data[column])
    label_encoders[column] = le

# Splitting the dataset into features (X) and target (y)
# Assuming 'Grade' as the target variable for this task
X = data.drop(['Grade'], axis=1)
y = data['Grade']

# Splitting the data into training and testing sets (75% train, 25% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Step 5 Machine Learning Part

#made by Yusuf İlteriş Kaan Günay
# Choosing the Decision Tree Classifier as it is one of the simpler models
dt_classifier = DecisionTreeClassifier(random_state=42)

# Performing 10-fold cross-validation on the training data
cv_scores = cross_val_score(dt_classifier, X_train, y_train, cv=10)

# Calculating average cross-validation score
average_cv_score = np.mean(cv_scores)

average_cv_score
```

```python
#made by Yusuf İlteriş Kaan Günay

# Training the decision tree classifier on the training set
dt_classifier.fit(X_train, y_train)

# Predicting the test set
y_pred = dt_classifier.predict(X_test)

# Calculating confusion matrix, accuracy, F1 score, and other metrics
conf_matrix = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='weighted')

# Calculating average accuracy, precision, recall
avg_accuracy = cross_val_score(dt_classifier, X_train, y_train, cv=5).mean()
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')

# Generating ROC curve values
try:
    y_prob = dt_classifier.predict_proba(X_test)
    fpr, tpr, thresholds = roc_curve(y_test, y_prob[:, 1], pos_label=1)
    roc_auc = auc(fpr, tpr)
except ValueError:
    roc_auc = None
    fpr, tpr = [None], [None]

# Displaying all the metrics with headers
print("Confusion Matrix:\n", conf_matrix)
print("\nAccuracy:", accuracy)
print("F1 Score:", f1)
print("Average Accuracy:", avg_accuracy)
print("Precision:", precision)
print("Recall:", recall)

if roc_auc is not None:
    print("ROC AUC:", roc_auc)
    print("False Positive Rate:", fpr)
    print("True Positive Rate:", tpr)
else:
    print("ROC curve values not applicable for Decision Trees")

    # i. Comparison of training and test data:
# Evaluate model performance metrics (accuracy, precision, recall, F1 score) on both datasets to assess generalization.
# Analyze overfitting (high training accuracy, low test accuracy) or underfitting (poor performance on both).
# Compare confusion matrices to understand error distribution.
# If applicable, review ROC/AUC for insights on class discrimination in training vs. test data.

```

```python

# Yusuf İlteriş Kaan Günay, Sultan Melda İnat, Mert Beşer, Yunus Kerem Saçar

#TASK 6 ENSEMBLE LEARNING

# Create classifiers
nb_classifier = GaussianNB()
knn_classifier = KNeighborsClassifier()
dt_classifier = DecisionTreeClassifier()
rf_classifier = RandomForestClassifier()
lr_classifier = LogisticRegression()

# Fit the classifiers
# Assuming X_train, y_train are already defined
nb_classifier.fit(X_train, y_train)
knn_classifier.fit(X_train, y_train)
dt_classifier.fit(X_train, y_train)
rf_classifier.fit(X_train, y_train)
lr_classifier.fit(X_train, y_train)

# Create an ensemble of all five classifiers
ensemble = VotingClassifier(estimators=[
    ('naive_bayes', nb_classifier),
    ('knn', knn_classifier),
    ('decision_tree', dt_classifier),
    ('random_forest', rf_classifier),
    ('logistic_regression', lr_classifier)
], voting='hard')

# Fit the ensemble classifier
ensemble.fit(X_train, y_train)

# Evaluate the ensemble classifier
ensemble_accuracy = ensemble.score(X_test, y_test)
print("Ensemble Accuracy:", ensemble_accuracy)
```

```python
# Yusuf İlteriş Kaan Günay, Sultan Melda İnat, Mert Beşer, Yunus Kerem Saçar

#TASK 7

# Feature Selection
selector = SelectKBest(score_func=f_classif, k='all')  # Adjust 'k' based on your needs
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)

# Train and evaluate all models with selected features
models = {
    'LogisticRegression': LogisticRegression(),
    'KNeighborsClassifier': KNeighborsClassifier(),
    'DecisionTreeClassifier': DecisionTreeClassifier(),
    'RandomForestClassifier': RandomForestClassifier(),
    'GaussianNB': GaussianNB()
}

best_model_name = None
best_accuracy = 0

for name, model in models.items():
    model.fit(X_train_selected, y_train)
    y_pred = model.predict(X_test_selected)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{name} Accuracy with selected features: {accuracy}")

    # Update the best model if the current model performs better
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_model_name = name

# Print the best model's name and accuracy
print(f"The best performing model is {best_model_name} with an accuracy of {best_accuracy:.4f}.")
```
